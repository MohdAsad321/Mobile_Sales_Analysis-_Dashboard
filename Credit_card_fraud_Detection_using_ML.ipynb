#**Importing Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset into pandas dataframe

credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# shape of the data
credit_card_data.shape
```

```python
# first five rows of the data

credit_card_data.head()
```

```python
# last five rows of the data

credit_card_data.tail()
```

```python
# data information
credit_card_data.info()
```

```python
# checking the missing values in each column.
credit_card_data.isnull().sum()
```

```python
# Dropping the rows which have missing values.

credit_card_data = credit_card_data.dropna()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data.shape
```

```python
# Distribution of legit and and fraudulent transactions.


credit_card_data['Class'].value_counts()
```

# This is highly imbalanced data.

# 0-----> normal transaction
# 1-----> fraud transaction

```python
# seperating the data for analysis.
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measure.
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare mean values for both transactions.

credit_card_data.groupby('Class').mean()
```

# Under Sampling
--> Build a sample dataset containing similar distribution of legit & fraud transacrtions

```python
legit_sample = legit.sample(n = 492)
```

# Concatenate two dataframes.

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Splitting the data into features and target.

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

# **Splitting the data into training & testing data.**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Model training**

```python
model = LogisticRegression(max_iter = 1000)
```

```python
# training the model with training data
```

```python
model.fit(X_train,Y_train)
```

# **Model Evaluation.**

```python
# accuracy score
```

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data is : ",training_data_accuracy )
```

```python
# accuracy score on testing data.

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on testing data is : ",testing_data_accuracy )
```
